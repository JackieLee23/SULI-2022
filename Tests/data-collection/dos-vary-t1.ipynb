```python
import matplotlib.pyplot as plt
import numpy as np
import os
root = "/home/jlee/ml-project/"
```

```python
dos_dict = {}
data_location = os.path.join(root, 'scba/test')
```

```python
for directory in os.listdir(data_location):
    if (directory[:6] == 'Params'):
        dos_file = os.path.join(data_location, directory, 'dos.dat')
        dos_dict[directory] = np.loadtxt(dos_file)
```

```python
def plot_one_dos(key, ax):
    
    ax.plot(dos_dict[key][:,0], dos_dict[key][:,1])

    ax.set_title(key[5:], fontsize = 16)
    
    ax.text(2, 0.8, key[7:], fontsize = 16)
```

```python
fig, ax = plt.subplots(11, figsize = (15, 55))

for i, key in enumerate(dos_dict.keys()):
    plot_one_dos(key, ax[i])
```
