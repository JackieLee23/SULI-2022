```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt
import os
from os import path
import sys
print(os.getcwd())
root = "/project/wyin/jlee/ml-project/"
util_loc = os.path.join(root, "utils")
data_loc = os.path.join(root, "inverse-shifted/data")
sys.path.append(util_loc)
from inverse_utilities import ScaledData
from sklearn.decomposition import PCA
```

```python
train_set = np.load(path.join(data_loc, "train-set.npz"))
train_dos = train_set['dos']
train_params = train_set['params']

pca = PCA(n_components=2)
principal_components = pca.fit_transform(train_dos)
print(principal_components.shape)
```

```python
def pca_color(ax, index, name):
    c = ax.scatter(principal_components[:,0], principal_components[:,1], c=train_params[:,index], s=1, cmap=plt.cm.get_cmap('RdPu'))
    
    plt.colorbar(c, ax = ax)
    ax.set_title(name)
```

```python
fig, ax = plt.subplots(1, 3, figsize =(15, 5))

pca_color(ax[0], 0, 't1')
pca_color(ax[1], 1, 't2')
pca_color(ax[2], 2, 'J')
```
