```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
root = "/project/wyin/jlee/ml-project"
util_loc = os.path.join(root, 'utils')
sys.path.append(util_loc)
data_loc = os.path.join(root, "dispersion-prediction/data")
from utilities import find_knn, ScaledData, see_results
```

```python
data = ScaledData(data_loc, X_name = "disp", y_name = "params")
print(data.train_y.shape)
```

```python
model = find_knn(data_loc, X_name = "disp", y_name = "params", scaled = True)
```

```python
def get_weights(distances):
    return distances ** -3

model = find_knn(data_loc, X_name = "disp", y_name = "params", scaled = True, weight_func = get_weights)
```

```python
predicted = model.predict(data.test_X)
percentiles = see_results(predicted, data.test_y, (5, 2), percentiles = np.arange(10), x_vals = np.arange(3), 
                          tick_labels = ['t1', 't2', 'j'])
```

```python
for i, (mse, predicted, truth) in enumerate(percentiles):
    print(f"Percentile {i}:")
    print(f"Truth: t1: {truth[0]}, t2: {truth[1]}, J: {truth[2]}")
    print(f"Predicted: t1: {predicted[0]:.4f}, t2: {predicted[1]:.4f}, J: {predicted[2]:.4f}")
    print()
```
