```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
root = "/project/wyin/jlee/ml-project"
util_loc = os.path.join(root, 'utils')
sys.path.append(util_loc)
data_loc = os.path.join(root, "dispersion-prediction/data")
from utilities import find_knn, ScaledData, see_results
```

```python
data = ScaledData(data_loc, X_name = "params", y_name = "disp")
print(data.train_y.shape)
```

```python
model = find_knn(data_loc, X_name = "params", y_name = "disp", scaled = True)
```

```python
def get_weights(distances):
    return distances ** -4

model = find_knn(data_loc, X_name = "params", y_name = "disp", scaled = True, weight_func = get_weights)
```

```python
predicted = model.predict(data.test_X)
percentiles = see_results(predicted, data.test_y, (5, 2), np.arange(10), np.linspace(0, 3, 193))
```
