```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
dos_dict = {}
```

```python
for directory in os.listdir('data'):
    if (directory.find('0.') != -1):
        dos_file = os.path.join('data', directory, 'dos.dat')
        dos_dict[directory] = np.loadtxt(dos_file)
```

```python
print("hi")
print(dos_dict.keys())
```

```python
def plot_dos(keys_list):
    plt.figure(figsize = (15, 5))

    for key in keys_list:
        plt.plot(dos_dict[key][:,0], dos_dict[key][:,1], label = key, linewidth = 0.7)

    plt.ylabel('dos')
    plt.xlabel('w')
    plt.legend()
```

```python
keys_list = [key for key in dos_dict.keys() if key[-8:] == '0.00-0.4']
plot_dos(keys_list)
```
