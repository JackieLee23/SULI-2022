```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import pytorch_lightning as pl
from pytorch_lightning import Trainer
from pytorch_lightning.loggers import TensorBoardLogger
import sys
import os

root = "/project/wyin/jlee/ml-project/dos-prediction"
util_loc = os.path.join(root, "utils")
data_loc = os.path.join(root, "data")
sys.path.append(util_loc)
from utilities import LitNeuralNet, DosDataModule
```

```python
class SchedLitNeuralNet(pl.LightningModule):
    def __init__(self, layer_sizes, lr = 0.01):
        super(SchedLitNeuralNet, self).__init__()
        
        modules = []
        for i in range(len(layer_sizes) - 1):
            modules.append(nn.Linear(layer_sizes[i], layer_sizes[i + 1]))
            
            if i != len(layer_sizes) - 2:
                modules.append(nn.ReLU())
        
        self.forward_prop = nn.Sequential(*modules)
        self.learning_rate = lr
        self.save_hyperparameters()
    
    def training_step(self, batch, batch_idx):
        params, dos = batch
        
        # Forward pass
        predicted = self.forward_prop(params)
        loss = F.mse_loss(predicted, dos)
        
        #log to tensorboard
        self.log("train_loss", loss)
        return loss
    
    def validation_step(self, batch, batch_idx):
        params, dos = batch
        
        # Forward pass
        predicted = self.forward_prop(params)
        loss = F.mse_loss(predicted, dos)
        
        #log to tensorboard
        self.log("val_loss", loss)
        return loss
        
    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.learning_rate)
        sch = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.2)
        return {
            "optimizer":optimizer,
            "lr_scheduler" : {
                "scheduler" : sch,
                "monitor" : "train_loss",
                
            }
        }
```

```python
batch_size = 1000
learning_rate = 0.001
#num_epochs = 50
dos_data = DosDataModule(data_loc, batch_size)
layer_sizes = [3, 128, 256, 512, 1024, 301]
logger = TensorBoardLogger('logs/nn_tests', name = f'scheduled-{learning_rate}, {batch_size}, {layer_sizes}, 0.7')


trainer = pl.Trainer(max_time = "00:00:30:00", logger = logger)
model = SchedLitNeuralNet(layer_sizes, lr = learning_rate)
trainer.fit(model, datamodule=dos_data)
```

```python
#Load from checkpoint
check_dir = "logs/nn_tests/scheduled-0.001, 1000, [3, 128, 256, 512, 1024, 301], 0.5/version_0/checkpoints"
for file in os.listdir(check_dir):
    print(file)
    check_file = os.path.join(check_dir, file)
    
print(check_file)

trainer = pl.Trainer(max_time="00:06:00:00", logger = logger)
model = SchedLitNeuralNet.load_from_checkpoint(checkpoint_path=check_file)
trainer.fit(model, datamodule = dos_data)
```
