```python
import numpy as np
import matplotlib.pyplot as plt
import sys
util_loc = "/project/wyin/jlee/ml-project/utils"
sys.path.insert(0, util_loc)
print(sys.path)
from utilities import order_validation
```

```python
results = order_validation("/project/wyin/jlee/ml-project/dos-prediction-full/neural-nets/hyperparameter-search/val-ends/linear-search-30-min")
```

```python
val_dict = {}
for pair in results:
    val_dict[pair[1]] = pair[0]
        
print(val_dict)
```

```python
mse_arr = np.zeros(20)

for i in range(20):
    mse_arr[i] = val_dict[f'Linear {i}']
    
plt.figure(figsize = (15, 5))
plt.yscale("log")
plt.ylabel("Validation loss after training 30 min")
plt.xlabel("Number of hidden layers")
plt.xticks(np.arange(20))
plt.plot(np.arange(20), mse_arr, label = "Linear ramps to 301 neurons")
plt.title("Linear ramp architectures")

plt.axhline(val_dict['[3, 87, 171, 255, 339, 423, 507, 301]'], color = "red", label = "6 layers, up to 507 neurons")
plt.axhline(val_dict['[3, 173, 343, 513, 683, 853, 1023, 301]'], color = "orange", label = "6 layers, up to 1020 neurons")

plt.legend()
```
